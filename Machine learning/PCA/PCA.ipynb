```python
import numpy as np


#文件读取
def loadDataSet(fileName,splitstr='\t'):
    stringArr=[]
    with open(fileName,'r') as fread:
        for line in fread.readlines():
            stringArr.append(map(float,line.strip().split(splitstr)))
    return np.mat(stringArr)
```

```python
dataSet=loadDataSet('testSet.txt')
```

```python
def pca(dataMat,dimNum):
    #中心化和标准化
    meanDataMat=np.mean(dataMat,axis=0)
    varDataMat=np.var(dataMat,axis=0)
    normalDataMat=np.divide((dataMat-meanDataMat),varDataMat)
    #计算协方差矩阵
    covMat=np.cov(normalDataMat,rowvar=False)
    #计算特征值和特征向量
    eigVals,eigVects=np.linalg.eig(covMat)
    #将特征值从大到小排序
    eigIndex=np.argsort(-eigVals)
    redEigVects=eigVects[:,eigIndex[:dimNum]]
    #转换数据空间(降维)
    lowDimData=normalDataMat*redEigVects
    #将降维后的投影数据转换到原空间
    reconData=np.multiply((lowDimData*redEigVects.T),normalDataMat)+np.mean(dataMat,axis=0)
    return lowDimData,reconData
```

```python
pca(dataSet,1)
```
